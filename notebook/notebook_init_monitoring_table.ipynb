```python
from pyspark.sql import SparkSession
from pyspark.sql.types import StructType, StructField, LongType, StringType, TimestampType, IntegerType, ShortType

```

```python
# Schema of the job_execution table
schema_job = StructType([
    StructField("job_id", LongType(), False),
    StructField("orchestration_id", LongType(), False),
    StructField("job_name", StringType(), False),
    StructField("job_type", StringType(), False),
    StructField("notebook_name", StringType(), False),
    StructField("source_name", StringType(), False),
    StructField("object_name", StringType(), False),
    StructField("job_start", TimestampType(), False),
    StructField("job_end", TimestampType(), True),
    StructField("valid_files", IntegerType(), False),
    StructField("invalid_files", IntegerType(), False),
    StructField("total_files", IntegerType(), False),
    StructField("source_layer", StringType(), False),
    StructField("target_layer", StringType(), False),
    StructField("target_table_name", StringType(), False),
    StructField("job_status", StringType(), False),
    StructField("job_message", StringType(), True),
    StructField("year", ShortType(), False),
    StructField("month", ShortType(), False),
    StructField("day", ShortType(), False)
])

df_job = spark.createDataFrame([], schema_job)
```

```python
table_name = "job_execution"

# Écriture de la table dans le dossier `Tables` de votre Lakehouse
df_job.write.format("delta").mode("overwrite").saveAsTable(table_name)
```
